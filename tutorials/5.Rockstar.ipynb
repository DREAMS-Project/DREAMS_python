```python
import numpy as np
```

```python
mvs = dreams_python.DREAMS('/standard/DREAMS') ## default is mass varied suite (CDM/varied_mass/SB6/)

run  = 4 ## load simulation number 4
snap = 90 ## load snapshot 90 (z=0)

rockstar_cat = mvs.read_rockstar(run, snap, keys=['Mvir', 'ID'])
fof_cat = mvs.read_group_catalog(run, snap, keys=['GroupMass'])

params, header = mvs.read_param_file('TNG_SB6.txt')

target_masses = params[:, 0]

rockstar_id = mvs.get_target_rockstar_index(run, snap, target_masses[run])
fof_id = mvs.get_target_fof_index(run, snap, target_masses[run])

rockstar_mass = np.log10(rockstar_cat["Mvir"])[rockstar_cat["ID"]==rockstar_id][0]
fof_mass = np.log10(fof_cat["GroupMass"][fof_id] * 1.00E+10/0.6909)

print( f'Target Halo Mass   : {target_masses[run]:0.3f}' )
print( f'FoF Halo Mass      : {fof_mass:0.3f}' )
print( f'Rockstar Halo Mass : {rockstar_mass:0.3f}')
```
